```python
#  importing libraries

import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Line Plot

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Bar plot

import pandas as pd

data = pd.DataFrame({
    "class":["A","B","c","D","E"],
    "student":[57,33,45,56,49]})

sns.barplot(x="class",y="student",data=data)
plt.title('Bar Plot')

# ploting
plt.show()
```

```python
# Count plot

sns.countplot(x='class',data=data)
plt.title("count plot")
plt.show()
```

```python
# Scatter plot

x=[1,2,3,4,5]       # connenction bet two variables
y=[5,2,7,8,4]
plt.scatter(x,y)
plt.show()
```

```python
# Hist
import numpy as np
data = np.random.rand(100)
plt.hist(data,bins=10)
plt.title("Histogram")
plt.show()
```

```python
# Box plot

sns.boxplot(y=data)             # for distribution and outlyer detecting
plt.title("Box Plot")           #There is single outlier present in data
plt.show()
```

```python
# Violin plot

# use box plot and KDE

sns.violinplot(y = data)
plt.title("violin plot")  # no outlier & normally distributed dataset
plt.show()
```

```python
# Pair plot

#connection between all fetures of dataframe

df =sns.load_dataset('iris')

sns.pairplot(df,hue ='species')
plt.show()
```

```python
# Heatmap

df =df.drop('species',axis =1)
correlation_matrix = df.corr()

correlation_matrix

sns.heatmap(df , annot=True ,cmap = 'coolwarm')
plt.title("Correlation Heatmap")
```

```python
# Pie chart

labels = ['A', 'B', 'C']
sizes = [30, 50, 20]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie chart")
plt.show()
```

```python
# Swarm plot

# Load the built-in Iris dataset
df = sns.load_dataset("iris")

# Create a swarm plot
sns.swarmplot(x='species', y='sepal_length', data=df)

# Add a title
plt.title("Swarm Plot of Sepal Length by Species")

# Show the plot
plt.show()
```
